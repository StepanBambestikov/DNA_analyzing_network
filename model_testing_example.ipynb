```python
import torch
import prediction_analyzer as pl
import pandas as pd
import excel_parameters as constant
import data_loader as loader
import network_classes as networks
```

First, load the trained network:

```python
nn_model = networks.SimpleNetwork()
nn_model.load_state_dict(torch.load("dH_model"))
```

And load out dataset:

```python
dataset = loader.get_dataset_from_excel_file("ML_Stepan.xlsx", label_column_number=constant.dH_column_number,
                                                 begin_feature_column=constant.begin_feature_column,
                                                 end_feature_column=constant.end_feature_column,
                                                 dna_to_numeric_strategy=None, first_row=constant.first_row)
dataloader = torch.utils.data.DataLoader(dataset=dataset, batch_size=dataset.__len__())
input_data, ground_truth_labels = next(iter(dataloader))
prediction = nn_model(input_data)
```

Lets draw histograms of prediction error, absolute and relative values, with 
prediction_analyzer module:

```python
error = pl.Graph('train error', prediction - ground_truth_labels)
normilized_error = pl.Graph('train normalized error', (prediction - ground_truth_labels) / ground_truth_labels)
pl.draw_error_histograms(error, normilized_error)
```

Now, lets draw correlation between prediction and truth labels, also with prediction_analyzer module:

```python
pair = pl.CorrelationPair(prediction, ground_truth_labels, 'train data')
pl.draw_correlation(pair)
```

Finally, save prediction and truth labels in some file, using any posible load function from pandas library, in our example lets use DataFrame.to_excel function

```python
pl.save_data_in_file(prediction, ground_truth_labels, 'saved_file.xlsx', pd.DataFrame.to_excel)
```
